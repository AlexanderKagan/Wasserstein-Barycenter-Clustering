```python
import numpy as np
import matplotlib.pyplot as plt
from CLUSTERING import clustering
from itertools import product
from time import  time
from UTILS import DiscreteDistrib, disc_gauss_generator
from tqdm import tqdm
```

```python
nsamples = 10

P1 = disc_gauss_generator(nsamples, 50, 3)
P2 = disc_gauss_generator(nsamples, 100, 3)
P3 = disc_gauss_generator(nsamples, 150, 3)
P4 = disc_gauss_generator(nsamples, 200, 3)
P5 = disc_gauss_generator(nsamples, 250, 3)
P6 = disc_gauss_generator(nsamples, 300, 3)

Ps = [P1, P2, P3, P4, P5, P6]
```

```python
K = 2
num_runs = 5

rho_range = [1e-4, 1e-3, 1e-2, 1e-1]
tau_range = [1, 5, 10, 25, 50]
times_dict = {}
labels_dict = {}

for tau, rho in tqdm(list(product(tau_range, rho_range))):
    times = []
    labels_list = []
    for run in range(num_runs):
        start = time()
        labels = clustering(Ps, K, 
                    centroid_update_kwargs={"rho": rho, "tau": tau, "nIter": 10000, "verbose_interval": 1000})
        end = time()
        
        times.append(end - start)
        labels_list.append(labels)
        
    times_dict[(tau, rho)] = times
    labels_dict[(tau, rho)] = labels_list
```
